```python
#Importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading dataset
df=pd.read_csv("/car_crashes.csv")
df.head
df
```

```python
#Scatter Plot
plt.scatter(df['alcohol'],df['not_distracted'])
plt.xlabel('alcohol')
plt.xlabel('not_distracted')
```

```python
#Bar Plot
plt.bar(df['speeding'],df['not_distracted'],width=0.2,color='green')
plt.xlabel('speeding')
plt.ylabel('not distracted')
```

```python
#Box Plots
plt.boxplot(df['speeding'])
```

```python
#Histograms
plt.hist(df['speeding'],width=0.7)
plt.xlabel('speeding')
plt.ylabel('frequency')
```

```python
#Boxenplot
sns.set_theme(style="whitegrid")
sns.boxenplot(df['alcohol'])
```

```python
#Relplot
sns.set_theme(style="white")
sns.relplot(sizes=(40, 400), alpha=.5, palette="muted",
            height=6,data=df['not_distracted'])
```

```python
#RegPlot
sns.set_theme(style="white")
sns.regplot(x=df['alcohol'],y=df['speeding'])
```
